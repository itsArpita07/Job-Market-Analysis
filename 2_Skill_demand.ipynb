# Most Demanded Skills for Top 3 Data Roles

```python
import pandas as pd
import  numpy as np
import seaborn as sns
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[(df['job_country'] == 'United States')].copy()
df_US
```

```python
df_skills = df_US.explode('job_skills')
df_skills[['job_title', 'job_skills']]
```

```python
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()
df_skills_count = df_skills_count.reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)
df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()
job_titles = sorted( job_titles[:3] )
job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1, figsize=(8, 6)) 
sns.set_theme(style="ticks")

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_count', y='job_skills', ax=ax[i], width=0.8) 
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].set_xticks(range(0, 45001, 5000))
    ax[i].legend().set_visible(False)

fig.suptitle('Top 5 Skills for Each Job Title', fontsize=16, weight='bold')
fig.tight_layout()
plt.show()
```

```python
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total')
df_job_title_count
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')
df_skills_perc['skill_percent'] = df_skills_perc['skill_count'] / df_skills_perc['jobs_total'] * 100 
df_skills_perc
```

```python
fig, ax = plt.subplots(len(job_titles), 1, figsize=(10, 6))

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], width=0.8)
    
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].legend().set_visible(False)
    ax[i].set_xlim(0, 80)

    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n, str(round(v, 2)) + '%', color='black', fontsize=9, va='center')

fig.suptitle('Likelihood of Skills Requested in Job Postings', fontsize=16, weight='bold')
fig.tight_layout()
plt.show()
```
